```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_xlabel('set x label')
axes.set_ylabel('set y label')
axes.set_title('graph')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_xlabel('set x label')
axes.set_ylabel('set y label')
axes.set_title('graph')

fig2 = plt.figure()
axes = fig2.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'g')
axes.set_xlabel('set x1 label')
axes.set_ylabel('set y1 label')
axes.set_title('graph1')

fig.show()
fig2.show()
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis2 = fig.add_axes([0.2,0.5,0.4,0.3])

axis1.plot(x,y,'b')
axis2.plot(y,x,'b')
```

```python
#Sub Plots
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Graph')
```

```python
fig, (axes1, axes2) = plt.subplots(1,2)
        
axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Graph')

axes2.plot(y,x,'r')
axes2.set_xlabel('x1')
axes2.set_ylabel('y1')
axes2.set_title('Graph1')
```

```python
fig, axes = plt.subplots(2,2)
axes[0][0].plot(x,y,'r')
axes[0][1].plot(x,y,'g')
axes[1][0].plot(x,y,'b')
axes[1][1].plot(x,y,'y')
```
